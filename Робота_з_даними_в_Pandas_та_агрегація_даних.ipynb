Використання [] для фільтрації вибірки

Фільтрація даних на основі умов:

```python
# Фільтрація атлетів, що отримали золоту медаль - loc method
gold_medals = df.loc[df['Medal'] == 'Gold']
print(gold_medals.head())
```

```python
# Фільтрація атлетів, що отримали золоту медаль - simple syntax
gold_medals = df[df['Medal'] == 'Gold']
print(gold_medals.head())
```

Фільтрація за кількома умовами (використовуйте оператори & (і) або | (або):)

```python
# Фільтрація атлетів, які отримали золоту медаль у чоловічих змаганнях
gold_men = df[(df['Medal'] == 'Gold') & (df['Gender'] == 'Men')]
print(gold_men.head())
```

```python
# Використання методу isin для перевірки умови на відповідність декільком значенням
gold_men = df[(df['Medal'].isin(['Gold','Silver'])) & (df['Gender'] == 'Men')]
print(gold_men.head())
```

```python
# Використання методу isin для перевірки умови на НЕвідповідність декільком значенням
gold_men = df[(~df['Medal'].isin(['Gold','Silver'])) & (df['Gender'] == 'Men')]
print(gold_men.head())
```

Застосування функцій агрегації на згрупованих DataFrames

```python
# Групування медалей за країною (NOC)
medals_by_noc = df.groupby('NOC', as_index=False, sort=False)['Medal'].count()
print(medals_by_noc)
```

```python
# Підрахунок мінімального і максимального місця на пʼєдесталі для кожної країни
agg_data = df.groupby('NOC')['Position'].agg(['min', 'max'])
print(agg_data)
```

```python
# Підрахунок кількості медалей та визначення унікальних видів спорту для кожної країни
agg_data = df.groupby('NOC').agg({
    'Medal': 'count',
    'Sport': 'nunique'
})
print(agg_data)
```

Застосування методу transform()

```python
# Додавання колонки з кількістю медалей для кожної країни (групування і трансформація)
df['Total_Medals'] = df.groupby('NOC')['Medal'].transform('count')
print(df.head())
```

Сортування в Pandas

```python
# Сортування атлетів за роком змагань та ім'ям
sorted_df = df.sort_values(by=['Year', 'Athlete Name'])
print(sorted_df.head())
```
