# 1. Matplotlib: Побудова найпростіших діаграм

```python
import pandas as pd
from pathlib import Path

if not Path("olympics_1896_2004.csv").exists():
  !wget https://github.com/jonfernandes/pandas_essential/raw/main/olympics_1896_2004.csv

filename = "olympics_1896_2004.csv"
print("Installed all of the necessary files for this section ...")
```

```python
df = pd.read_csv(filename, skiprows=5)
df.head()
```

```python
import matplotlib.pyplot as plt
```

Приклад 1: Побудова лінійного графіка

```python
# Підрахунок кількості медалей по роках
medals_per_year = df.groupby('Year')['Medal'].count()
medals_per_year.head()
```

```python
medals_per_year.plot()
```

```python
# Побудова лінійного графіка
plt.figure(figsize=(10, 6))
plt.plot(medals_per_year, marker='o')
plt.title('Кількість медалей по роках')
plt.xlabel('Рік')
plt.ylabel('Кількість медалей')
plt.grid(True)
plt.show()
```

Приклад 2: Побудова стовпчастої діаграми

```python
# Підрахунок кількості медалей для кожної країни (NOC)
medals_per_country = df['NOC'].value_counts().head(10)  # Топ-10 країн
medals_per_country
```

```python
# Побудова стовпчастої діаграми
plt.figure(figsize=(10, 6))
medals_per_country.plot(kind='bar')
plt.title('Топ-10 країн за кількістю медалей')
plt.xlabel('Країна')
plt.ylabel('Кількість медалей')
plt.xticks(rotation=45)
plt.show()
```

```python
# Побудова стовпчастої діаграми з горизонтальними табами
plt.figure(figsize=(10, 6))
medals_per_country.plot(kind='barh')
plt.title('Топ-10 країн за кількістю медалей')
plt.xlabel('Кількість медалей')
plt.ylabel('Країна')
plt.xticks(rotation=45)
plt.show()
```

# 2. Налаштування відображення діаграм в Matplotlib

Приклад 3: Налаштування кольорів і стилю

```python
# Побудова лінійного графіка з додатковим налаштуванням
plt.figure(figsize=(10, 6))
plt.plot(medals_per_year, color='green', marker='s', linestyle='--', linewidth=2)
plt.title('Кількість медалей по роках', fontsize=14)
plt.xlabel('Рік', fontsize=12)
plt.ylabel('Кількість медалей', fontsize=12)
plt.grid(True, linestyle='-.')
plt.show()
```

Приклад 4: Додавання легенди і сітки

```python
# Побудова графіка з кількома лініями
medals_by_gender = df.groupby(['Year', 'Gender'])['Medal'].count().unstack()

plt.figure(figsize=(10, 6))
plt.plot(medals_by_gender['Men'], label='Чоловіки', color='blue', marker='o')
plt.plot(medals_by_gender['Women'], label='Жінки', color='orange', marker='^')
plt.title('Кількість медалей по роках за статтю')
plt.xlabel('Рік')
plt.ylabel('Кількість медалей')
plt.legend()
plt.grid(True)
plt.show()
```

# 5. Побудова просунутих візуалізацій у Seaborn

Приклад 7: Створення boxplot для порівняння медалей по країнах

```python
medals_over_year = df.groupby(['Year', 'NOC']).size().reset_index(name='Medal_Count')
medals_over_year = medals_over_year[medals_over_year['NOC'].isin(medals_per_country.index)]
medals_over_year.head()
```

```python
# Побудова boxplot для порівняння країн за кількістю медалей
plt.figure(figsize=(12, 6))
sns.boxplot(x='NOC', y='Medal_Count', data=medals_over_year)
plt.title('Порівняння кількості медалей у топ-10 країн')
plt.xlabel('Країна')
plt.ylabel('Кількість медалей')
plt.xticks(rotation=45)
plt.show()
```

```python
medals_over_year[medals_over_year['NOC'] == 'GBR'].sort_values(by='Medal_Count', ascending=False).head(5)
```

Приклад 8: Створення heatmap для аналізу медалей за роками і країнами


```python
# Підрахунок кількості медалей за країнами і роками
medals_pivot = df.pivot_table(index='NOC', columns='Year', values='Medal', aggfunc='count', fill_value=0)
medals_pivot = medals_pivot[medals_pivot.index.isin(medals_per_country.index)]
medals_pivot
```

```python
# Побудова теплової карти (heatmap)
plt.figure(figsize=(12, 8))
sns.heatmap(medals_pivot, cmap='Blues')
plt.title('Кількість медалей за країнами і роками')
plt.show()
```

```python
def preprocess(filename = "olympics_1896_2004.csv"):
  """Preparing and transforming dataframe"""
  print(f"Preprocessing {filename} ...\n")
  ordered_medals = pd.api.types.CategoricalDtype(categories=["Bronze", "Silver", "Gold"], ordered=True)
  dtype_mapper = {"Year": "int64",
                "City": "string",
                "Sport": "string",
                "Discipline": "string",
                "Athlete Name": "string",
                "NOC": "string",
                "Gender": "category",
                "Event": "string",
                "Event_gender": "category",
                "Medal": ordered_medals}
  df = (pd.read_csv(filename, skiprows=5, dtype=dtype_mapper)
        .drop('Position', axis=1)
  )
  df["Event Gender"] = df["Event Gender"].astype("category")
  df.loc[24676, "Gender"] = "Women"
  df.Sport = df.Sport.str.lower()
  df.Discipline = df.Discipline.str.lower()
  df.Event = df.Event.str.lower()
  df.NOC = df.NOC.str.upper()
  return df
```

```python
df = preprocess()
df.head()
```
